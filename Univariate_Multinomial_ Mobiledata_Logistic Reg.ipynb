```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectKBest # K= number of required feature
from sklearn.feature_selection import chi2  # Chi will discover value of each x related to y
```

```python
df=pd.read_csv("E:/Neha Python/Machin Learning Notes/train2.csv")
```

```python
df
```

```python
x=df.iloc[:,0:20]# independant columns
x
```

```python
y=df.iloc[:,20]
y
```

```python
# Univariate Selection
# Apply SelectKBest class to extract top 10 best features
bestfeatures=SelectKBest(score_func=chi2,k=20)
fit=bestfeatures.fit(x,y)
dfscores=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(x.columns)

# concat two dataframes for better visualization

featureScores=pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns=['Specs','Score']   # namimg the dataframe columns
print(featureScores.nlargest(20,'Score'))   # print 20 best features (can vary value 20 eith required fields)

# Thus it gives the x fields (like 20 fields,10 fields mentioned in print statement) having maximum score to minimum
```

```python
print(featureScores.nlargest(10,'Score'))   # print 10 best features (can vary value 10 eith required fields)
#{For cheking purpose}
```

```python
# now how many fields out of 20 we have to choose is dependant on checking accuracy with variations in field
#(Like check accuracy for 5 fields , then check for 10 fields , which is more accordingly we have to delete or add fields)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
import matplotlib.pyplot as plt
model= ExtraTreesClassifier()
model.fit(x,y)
```

```python
# Feature Importance
print(model.feature_importances_) # use inbuilt class feature_importances
# Plot the graph of features for better visualization

feat_importances=pd.Series(model.feature_importances_,index=x.columns)
feat_importances.nlargest(15).plot(kind='bar')  # instead of 'bar' if we take 'barh' graph will be horizontal
plt.show()
```

```python
# Correlation Matrix with Heatmap

import seaborn as sns
# get correlation of each features in dataset

corrmat= df.corr() # preparing corralation for rows
top_corr_features=corrmat.index
plt.figure(figsize=(25,25))

# plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap='RdYlGn')  ## preparing corralation for columns 
#annot is used to show the values in boxes
```

```python
# #Apply model for best 4 features selected from above feature selection procedure

x2=df[['battery_power','ram','px_height','px_width']]
x2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# random state=0 means it starts from index 0
(x_train,x_test,y_train,y_test) = train_test_split(x2, y, test_size=25, random_state=0)
```

```python
model = LogisticRegression()
```

```python
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred =classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
