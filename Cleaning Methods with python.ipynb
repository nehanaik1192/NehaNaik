```python
# Making list of missing values (like N/a, n/a,-- etc)
missing_values=["n/a","na","--"]
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/dataoptimal/posts/master/data%20cleaning%20with%20python%20and%20pandas/property%20data.csv",na_values=missing_values)
```

```python
df  # Now it will remove n/a , na, -- type if entries here to give NaN
```

```python
# "Exceptional Handling" 
# In OWN_OCCUPIED column we have to replace any numeric value with string rather with NaN

count=0 # taking anything=0 means taking counter variable
for row in df['OWN_OCCUPIED']:
    try:
        int(row)                                   # Means if the row is integer then replace that with nan. other wise pass it 
        df.loc[count,'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    count+=1
```

```python
df.OWN_OCCUPIED # Now 12 is replaced by NaN
```

```python
df.NUM_BATH
```

```python
# "Exceptional Handling" 
# In OWN_OCCUPIED column we have to replace any numeric value with string rather with NaN

count=0 # taking anything=0 means taking counter variable
for row in df['NUM_BATH']:
    try:
        float(row)                                # Means if the row is integer then replace that with nan. other wise pass it 
        pass
    except ValueError:
        df.loc[count,'NUM_BATH']=np.nan   
    
    count+=1
```

```python
df = df.replace('HURLEY', np.nan)
```

```python
df
```

```python
df.head(10)  # In OWN_OCUPIED the values are Y and N Yes/No so we will think according to other values. Here we will fill NaN 
# with Y in Barklay and Washington NaN with Y
```

```python
df['ST_NUM']=df['ST_NUM'].fillna(df.ST_NUM.median())
```

```python
df['ST_NUM']
```

```python
df['NUM_BEDROOMS']=df['NUM_BEDROOMS'].fillna(df.NUM_BEDROOMS.median())
```

```python
df['NUM_BEDROOMS']
```

```python
df['NUM_BATH']=df['NUM_BATH'].fillna(df.NUM_BATH.median())
```

```python
df1=df.drop('PID',axis=1)
```

```python
df1['OWN_OCCUPIED']=df['OWN_OCCUPIED'].fillna(df1.OWN_OCCUPIED.mode())  # As it is string value Y / N so take mode in this case
```
