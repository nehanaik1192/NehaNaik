```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectKBest # K= number of required feature
from sklearn.feature_selection import chi2  # Chi will discover value of each x related to y
```

```python
df=pd.read_csv("E:/Neha Python/Machin Learning Notes/train2.csv")
```

```python
df
```

```python
x=df.iloc[:,0:20]# independant columns
x
```

```python
y=df.iloc[:,20]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Feature Extraction:
test=SelectKBest(score_func=chi2, k=4)  # k=4 means select 4 best field
fit=test.fit(x,y)   
```

```python
#Summarize scores
np.set_printoptions(precision=3) # precision 3 means again 3 best fields are chosen

print(fit.scores_)


# now here it will shoe the p values for all 8 fields. And from that it will select the fields having maximum values 
#with descending order. Now in this max values are 2175.565,1411.887 ,181.304, 111.52  are the max value 4 fields are selected 
# now see index of those fields in table. It will be glucose, insuline, etc....
```

```python
features=fit.transform(x) # Due to precision =3 it is giving only 3 columns out of best 4
print(features)
```

```python
# summarize selected features
print(features[0:5,:])
```

```python
# Import your necessary dependencies

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
# Feature Extraction
model=LogisticRegression()
rfe=RFE(model,3) # Again select 3 best fildes (or take number of fields how many required)
fit=rfe.fit(x,y)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_)) # will give true for field which is selected
print("Feature Ranking: %s" % (fit.ranking_)) # Gives rank according to selection (by p value) 
#It will give rank 1 which is to be selected
```
